**Question 7.** Let $X^k_i ∼ \text{Ber}(p)$ be the $i$th sample from $k$th batch. In each batch, we collect $n$ samples. We assume $X^k_i$ and $X^k_j$ are independent with each other for different $i$ and $j$. Let $S_k = X^k_1 + \cdots + X^k_n$ for $k = 1, \cdots , m$, which is the sum of $n$ Bernoulli random variables. Now, let $p = 0.4$, $n = 500$, and simulate $m = 1000$ different $S_k$.

a. Make a histogram for $S_j$.

b. Let $Y_k ∼ \mathcal{N}(\mu, \sigma^2)$. Find out the best $\mu$ and $\sigma^2$ and simulate $m = 1000$ of $Y_j$ which has a similar histogram as in (a).

***

### Solution

We are given:

$X^k_i ∼ \text{Ber}(p)$, $p=0.4$

Each batch $k$ consists of $n=500$ independent samples $X^k_i$.

$S_k = X^k_1 + \cdots + X^k_n$ for each batch $k$. This is the sum of $n$ Bernoulli random variables.

Consider $k \in \{m\in\mathbb{N} : 1\leq m\leq 1000\}$.

#### Part (a) - Plot a histogram

First, we simulate $S_k$.

Since $S_k$ is the sum of $n$ independently and identically distributed Bernoulli random variables with probability $p=0.4$. $S_k$ follows a binomial distribution. We have 500 samples for each batch $k$. Thus,

$$S_k ∼ \text{Bin}(n,p)$$

Expectation $E(S^k)=n\cdot p = 500 \times 0.4 = 200$

Variance $\text{Var}(S_k) = n\cdot p\cdot(1-p) = 500 \times 0.4 \times (1-0.4) = 500\times0.4\times0.6=120$

We then generate the histogram for $S_k$. We do this for 1000 batches, or $m=1000$.

```python
# Setup: Importing libraries and modules

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Setting parameters

p = 0.4  # probability of a success
n = 500  # samples per batch
m = 1000 # batches

# Simulating m batches of S_k ~ Binomial(n, p)
S_k = np.random.binomial(n, p, m)

# Plotting histogram for S_k
plt.hist(S_k, bins=30, alpha=0.7)
plt.xlabel('$S_k$')
plt.ylabel('Frequency')
plt.title('Histogram of $S_k$')
plt.show()
```

#### Part (b) - Find out the best values of $\mu$ and $\sigma^2$ to get a similar histogram

Let's find the theoretical and experimental values of $\mu$ and $\sigma^2$.

To approximate the distribution of $S_k$ with a normal distribution, we use the Central Limit Theorem (CLT), which suggests that $S_k$ can be approximated by a normal distribution with:

The theoretical values are as per the formulas above.

We can find the experimental values this way:

$$\mu = \text{mean}(S_k)$$
$$\sigma^2 = \text{std}(S_k)$$


```python
# Parameters for the normal approximation, theoretical
mu = n * p
var = n * p * (1 - p)
sigma = np.sqrt(var)

# Parameters for the normal approximation, experimental
mu_exp = np.mean(S_k)
sigma_exp = np.std(S_k)

# Simulating Y_j ~ N(mu, sigma^2), theoretical
Y_k_t = np.random.normal(mu, sigma, m)

# Simulating Y_j ~ N(mu, sigma^2), experimental
Y_k_e = np.random.normal(mu_exp, sigma_exp, m)

# Plot histogram for Y_k, theoretical
plt.hist(Y_k_t, bins=30, alpha=0.7)
plt.xlabel('$Y_k$')
plt.ylabel('Frequency')
plt.title('Histogram of $Y_k$')
plt.show()

# Plot histogram for Y_k, experimental
plt.hist(Y_k_e, bins=30, alpha=0.7)
plt.xlabel('$Y_k$')
plt.ylabel('Frequency')
plt.title('Histogram of $Y_k$')
plt.show()
```
